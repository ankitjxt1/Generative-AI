```python
import numpy as np
```

```python
#array
oned_array = np.array([1, 2, 3])
print(oned_array)
```

```python
rank1_vector = np.array([1, 2, 3, 4])
print(rank1_vector)
```

```python
#ndim--find the dimension of the vector
print(rank1_vector.ndim)
```

```python
#size--total number of item
print(rank1_vector.size)
```

```python
print(rank1_vector.transpose())
```
